```python
"""
2.Data preprocessing
Categories variables need to be transformed into numeric variables
Transform the embarkment port
There are three ports: C= Cherbourg, Q=Queenstown, S=Southampton
"""
```

```python
titanic.Embarked.unique()
```

```python
ports=pd.get_dummies(titanic.Embarked, prefix='Embarked')
ports.head()
```

```python
titanic=titanic.join(ports)
titanic.drop(['Embarked'],axis=1,inplace=True)
```

```python
titanic.head(10)
```

```python
titanic.Sex=titanic.Sex.map({'male':0,'female':1})
```

```python
"""
Extract the target variable
Create an X dataframe with the input features and an y series with the target (Survived)"""
```

```python
y=titanic.Survived.copy() #copy y columns values out, target variable _ bien phu thuoc
x=titanic.drop(['Survived'], axis=1) # then, drop y column _ bien doc lap
```

```python
"""
Drop not so important features
For the first model, we ignore some categorical features which will not add to much of a signal
"""
```

```python
x.drop(['Cabin'], axis=1, inplace=True)
x.drop(['Ticket'], axis=1, inplace=True)
x.drop(['Name'], axis=1, inplace=True)
x.drop(['PassengerId'], axis=1, inplace=True)
```

```python
x.info()
```

```python
"""
All features are now numeric, ready for regression and Missing value treatment
"""
```

```python
x.isnull().values.any()
```

```python
x[pd.isnull(x).any(axis=1)]
```

```python
x.Age.fillna(int(x.Age.mean()), inplace=True)
```

```python
x.isnull().values.any()
```

```python
#Split dataset to train and test set
from sklearn.model_selection import train_test_split
    #80% go into the training test, 20% in the validation test
x_train, x_valid, y_train, y_valid = train_test_split(x,y,test_size=0.2, random_state=7)
```

```python
#3.using the logistics regression module in SciKitLeanrn
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train) #fit the model
```

```python
#4. Evaluation the model _ danh gia mo hinh
model.score(x_train, y_train)
```

```python
"""
The score on the training set is much better than on the validation set,
and indication that could be overfitting and not being a general model.
"""
model.score(x_valid, y_valid)
```
