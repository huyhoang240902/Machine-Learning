```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler   #for scaling data
from sklearn.cluster import KMeans #for K-mean Cluster
```

```python
data=pd.read_csv('C:/Users/Admin/Downloads/17_Dang Pham Huy Hoang_Ch3_NBC.csv')
data
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
data['fixed acidity'].fillna(data['fixed acidity'].mean(), inplace=True)
data['volatile acidity'].fillna(data['volatile acidity'].mean(), inplace=True)
data['citric acid'].fillna(data['citric acid'].mean(), inplace=True)
data['residual sugar'].fillna(data['residual sugar'].mean(), inplace=True)
data['chlorides'].fillna(data['chlorides'].mean(), inplace=True)
data['pH'].fillna(data['pH'].mean(), inplace=True)
data['sulphates'].fillna(data['sulphates'].mean(), inplace=True)
```

```python
data = data.drop(data[data.type == 'white'].index)
```

```python
data['type'].unique()
```

```python
data.corr()
```

```python
inputs=data[['fixed acidity','volatile acidity','citric acid','sulphates','alcohol']]
target=data.quality
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
